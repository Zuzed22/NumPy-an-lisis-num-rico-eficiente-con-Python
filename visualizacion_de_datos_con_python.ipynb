# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

Conociendo los datos

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980,2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

```python
col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
```

```python
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

Creando el primer gráfico con matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de Colombianos hacia Canadá')
plt.xlabel('Años')
plt.ylabel('Número de Inmigrantes')
plt.show()
```

EJERCICIO

```python
df_comparacion = df.loc[['Brasil', 'Argentina'], anos]
```

```python
df_comparacion = df_comparacion.T
```

```python
df_comparacion.head()
```

```python
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

Añadiendo títulos y rótulos

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de Colombianos hacia Canadá\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

Creando subplots en dos direcciones #1

```python
fig, axs = plt.subplots(1,2,figsize=(10,3))

axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de Colombianos hacia Canadá\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración Colombianos\nhacia Canadá 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.show()
```

```python
datos_col.describe()
```

Creando subplots en dos direcciones #2

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigración de los 4 mayores países sudamericanos\nhacia Canadá desde 1980 hasta 2013')

axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid()

plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
df = pd.DataFrame(ventas_2022, index=tiendas)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
plt.subplots_adjust(wspace=0.3, hspace=0.4)
fig.suptitle('Ventas en el período de enero a diciembre de 2022 en las tiendas A, B, C y D')

axs[0, 0].plot(df.loc['A'])
axs[0, 0].set_title('Ventas en la tienda A')
axs[0, 1].plot(df.loc['B'])
axs[0, 1].set_title('Ventas en la tienda B')
axs[1, 0].plot(df.loc['C'])
axs[1, 0].set_title('Ventas en la tienda C')
axs[1, 1].plot(df.loc['D'])
axs[1, 1].set_title('Ventas en la tienda D')

for ax in axs.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('Número de ventas')
    ax.grid()

plt.show()
```

Añadiendo anotaciones

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')

# Now, the 'Region' column exists in df, and the following code will work
sudamerica = df.query('Region == "America del Sur"')
sudamerica.head(10)
```

```python
sudamerica.set_index('Region', inplace=True)
sudamerica.head()
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
           'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,3))
ax.bar(sudamerica.index, sudamerica['Total'],color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo de 1980 hasta 2013',
             loc='left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
           'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,3))
ax.bar(sudamerica.index, sudamerica['Total'],color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo de 1980 hasta 2013',
             loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)

colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
           'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,3))
ax.bar(sudamerica_sorted.index, sudamerica_sorted['Total'],color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo de 1980 hasta 2013',
             loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores =[]
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize=(10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y', which='both', size=0)

ax.bar(sudamerica_sorted.index, sudamerica_sorted['Total'],color=colores)
ax.set_title('Sudamérica: es el segundo país más migrantes\nhacia Canadá durante el periodo 1980 - 2013',
             loc='left', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=10)
# ax.xaxis.set_tick_params(labelsize=12)
# ax.yaxis.set_tick_params(labelsize=12)

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j), color='black', fontsize=8.5, ha='left', va='center')

fig.savefig('/content/inmigrantes_sudamericanos_canada.png',
            transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```
